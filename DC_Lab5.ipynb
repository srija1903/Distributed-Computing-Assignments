# BULLY ALGORITHM

```python
def bully_algorithm(failed,elec,n):
  active_nodes = []
  #storing all the active nodes
  for i in range(elec,n):
    if(failed[elec-1] == 0):
      active_nodes.append(i)
  #declaring a variable to store a coordinator
  global cdr
  if(len(active_nodes) > 0):
    print("\n")
    #Sending message from the node that initiates election to all the nodes
    for i in range(elec,n):
      print("Message sent from node ",elec,"to node ",i+1)
    
    #Sending the responses from the active nodes
    for i in range(elec,n):
      if(failed[i] == 0):
        print("Response sent from node ",i+1,"to ",elec)
        cdr = i+1
    #Once the process is completed one active node is removed from the list
    active_nodes.pop(0)
    if(len(active_nodes) > 0):
      print(active_nodes," are the nodes which are active")
    bully_algorithm(failed,elec+1,n)

n = int(input("Enter number of nodes:"))
nodes = []
failed = [] #Stores the status of each node. Default:-active
for i in range(1,n+1):
    nodes.append(i)
    failed.append(0)

#Considering the count of the number of the nodes one wishes to crash
fail_count = int(input("Enter the number of crashed nodes.(Write 0 if none):"))

if(fail_count == 1):
    node = int(input("Enter the ID of the crashed node:"))
    print("Node ",node," is crashed")
    failed[node-1] = 1
   
if(fail_count > 1):
  failed_nodes = [] #stores the list of the al the nodes one would crash
  for i in range(0,fail_count):
    node = int(input("Enter the ID of the crashed node:"))
    failed_nodes.append(node)
    print("Node ",node," is crashed")
  for i in failed:
    failed[failed_nodes[i]-1] = 1 #changes the status of the crashed node to 1

elec = int(input("Enter the node that initiates election:"))
#checking if the node is crashed or not
if(failed[elec-1] == 1):
    print("Sorry we can't proceed further as the node is crashed")
else:
  bully_algorithm(failed,elec,n)
  #The coordinator announces itself
  for i in nodes:
    print(cdr," announces itself as a coordinator to ",i)
```
